# Build a neural network 


```python
# ill start by defining the structure of it 
n = 2 # the number of inputs 
num_hidden_layers = 2
m = [2,2] # the number of nodes in each layer 
num_nodes_output = 1 
```

```python
# ill implement this only on numpy as a start 
import numpy as np 

num_nods_prevlayer = n 

network = {} 

## intuition: loop through each layer ad randomly intialize the weight and biases associated with each node 
# we are going to add 1 to the number of hidden layers in order to include the outpit layer

for layer in range (num_hidden_layers + 1): 

    if layer == num_hidden_layers:
        layer_name = 'output'
        num_nodes = num_nodes_output
    
    else: 
        layer_name = f'layer_ {layer + 1}'
        num_nodes = m[layer] 


    # intialization of the weights and biases for the current layer 
    network [layer_name] = {}
    for node in range (num_nodes):
        node_name = f'node_{node + 1}'
        network[layer_name][node_name] ={
            'weights': np.around(np.random.uniform(size= num_nods_prevlayer), decimals=2),
            'biases': np.around(np.random.uniform(size= num_nods_prevlayer), decimals=2)
        } 

    num_nods_prevlayer = num_nodes

print(network)




```
